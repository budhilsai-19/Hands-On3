```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Function f(n) with nested loops
def f(n):
    return 1 + sum(1 for i in range(n) for j in range(n))

# Measure time for different n values
n_values = range(1, 1001, 50)
times = []
for n in n_values:
    start = time.time()
    f(n)
    times.append(time.time() - start)

# Plot time vs. n
plt.plot(n_values, times, 'bo-', label='Measured Time')

# Fit a 2nd-degree polynomial (lower bound)
coefficients_lower = np.polyfit(n_values, times, 2)
poly_lower = np.poly1d(coefficients_lower)
plt.plot(n_values, poly_lower(n_values), 'r--', label='Lower Bound (2nd degree)')

# Fit a 3rd-degree polynomial (upper bound)
coefficients_upper = np.polyfit(n_values, times, 3)
poly_upper = np.poly1d(coefficients_upper)
plt.plot(n_values, poly_upper(n_values), 'g--', label='Upper Bound (3rd degree)')

# Add labels and legend
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Time vs. n for f(n)')
plt.legend()
plt.grid(True)
plt.show()

# Print polynomial coefficients for bounds
print("Lower Bound (2nd-degree polynomial) coefficients:", coefficients_lower)
print("Upper Bound (3rd-degree polynomial) coefficients:", coefficients_upper)

```

Big-O: This is often approximated by the polynomial with the highest degree that fits the upper bound. In this case, it's the 3rd-degree polynomial.

Big-Omega: This is often approximated by the polynomial with the degree that fits the lower bound. In this case, it's the 2nd-degree polynomial.

Big-Theta: If the polynomial fitting for the lower and upper bounds have the same degree, you can approximate Big-Theta to be that degree polynomial. In this case, it's expected to be quadratic, but you need to check for consistency in practice.


```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Function f(n) with nested loops
def f(n):
    return 1 + sum(1 for i in range(n) for j in range(n))

# Measure time for different n values
n_values = range(1, 1001, 50)
times = []
for n in n_values:
    start = time.time()
    f(n)
    times.append(time.time() - start)

# Fit a 2nd degree polynomial and plot the fitted curve
coefficients = np.polyfit(n_values, times, 2)
poly = np.poly1d(coefficients)

# Plot time vs. n and the fitted polynomial
plt.figure(figsize=(12, 6))
plt.plot(n_values, times, 'bo-', label='Time')
plt.plot(n_values, poly(n_values), 'r--', label='Fitted Polynomial')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Time vs. n for f(n)')
plt.legend()
plt.grid(True)
plt.show()

# Zoom in on the region where deviation might occur
# Adjust the range based on your observations
n_zoom = range(150, 300, 10)
times_zoom = [times[n_values.index(n)] for n in n_zoom]
poly_zoom = poly(n_zoom)

# Plot the zoomed-in region
plt.figure(figsize=(12, 6))
plt.plot(n_zoom, times_zoom, 'bo-', label='Time (Zoomed In)')
plt.plot(n_zoom, poly_zoom, 'r--', label='Fitted Polynomial (Zoomed In)')

# Identify and annotate deviation point
deviation = np.abs(np.array(times_zoom) - poly_zoom)
deviation_index = np.argmax(deviation)
n_0 = n_zoom[deviation_index]

plt.axvline(x=n_0, color='g', linestyle='--', label=f'n_0  {n_0}')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Zoomed-In Plot of Time vs. n for f(n)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Original function f(n)
def f_original(n):
    return 1 + sum(1 for i in range(n) for j in range(n))

# Modified function f(n)
def f_modified(n):
    x = 1
    y = 1
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            x += 1
            y = i + j
    return x

# Measure time for different n values
n_values = range(1, 1001, 50)
times_original = []
times_modified = []

for n in n_values:
    # Time original function
    start = time.time()
    f_original(n)
    times_original.append(time.time() - start)

    # Time modified function
    start = time.time()
    f_modified(n)
    times_modified.append(time.time() - start)

# Plot time vs. n for both functions
plt.figure(figsize=(12, 6))
plt.plot(n_values, times_original, 'bo-', label='Original Function')
plt.plot(n_values, times_modified, 'go-', label='Modified Function')

# Fit 2nd degree polynomials for both functions
coefficients_original = np.polyfit(n_values, times_original, 2)
coefficients_modified = np.polyfit(n_values, times_modified, 2)
poly_original = np.poly1d(coefficients_original)
poly_modified = np.poly1d(coefficients_modified)

# Plot fitted curves
plt.plot(n_values, poly_original(n_values), 'b--', label='Fitted Polynomial (Original)')
plt.plot(n_values, poly_modified(n_values), 'g--', label='Fitted Polynomial (Modified)')

# Add labels and legend
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.title('Time vs. n for Original and Modified Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Finding the mid of the array
        L = arr[:mid]  # Dividing the elements into 2 halves
        R = arr[mid:]

        merge_sort(L)  # Sorting the first half
        merge_sort(R)  # Sorting the second half

        i = j = k = 0

        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

    return arr

# Test the function
test_array = [5, 2, 4, 7, 1, 3, 2, 6]
sorted_array = merge_sort(test_array)
print("Sorted array:", sorted_array)
```
